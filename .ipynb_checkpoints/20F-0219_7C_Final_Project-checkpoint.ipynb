#### -Data Cleaning:
We opted for the approach of replacing empty strings with pd.NA in the DataFrame before checking for null values using isnull().sum() because pandas may not inherently recognize empty strings as null or missing values during standard null value checks. By explicitly marking empty strings as pd.NA, we ensure that these values are correctly identified as missing when calculating null value counts. This approach enhances the accuracy of our analysis by treating both empty strings and other conventional missing value representations uniformly, providing a comprehensive assessment of null values in the dataset.

In our analysis, we started by addressing missing or hidden values in the dataset. To achieve this, we utilized the Pandas library in Python. Specifically, we replaced empty strings and the term 'Hidden' with `NaN` (Not a Number) throughout the entire DataFrame.

The reason for replacing hidden with `NaN` as to us we considered it unknown data and delt with it based upon qualitative and quantitative data.

```python
# Replace empty and Hidden strings with NaN in the entire DataFrame
_raw_data.replace('', pd.NA, inplace=True)
_raw_data.replace('Hidden', pd.NA, inplace=True) # later on we will be dealing them

# Display the count of missing values in each column
for col in _raw_data.columns:
    count_nan = _raw_data[col].isna().sum()
    print(f'{col}: {count_nan} missing values', end='\n\n')
```

###### Understanding the Distribution

In this crucial phase of our analysis, we delved into the distribution of data that was previously unknown or undisclosed to us. Leveraging visualizations such as histograms, kernel density plots, and box plots, we gained valuable insights into the underlying patterns and characteristics of the data.

The significance of this exploration lies in our ability to make informed decisions regarding the treatment of unknown values. By carefully examining the distribution graphs, we were empowered to make strategic choices on whether to impute missing values using statistical measures such as median, mode, or mean, taking into consideration the skewness observed in the data.

This was only for the Quantitative data.

```python
df = _raw_data.copy()
fig, axes = plt.subplots(nrows=3, ncols=3, figsize=(15, 10))
axes = axes.flatten()
# List of column names for which we are checking
variables = ['Age', 'Purchase_Amount', 'Average_Spending_Per_Purchase', 'Purchase_Frequency_Per_Month', 'Brand_Affinity_Score', 'Month', 'Year']

for i, variable in enumerate(variables):
    # Replace non-numeric values with NaN
    df[variable] = pd.to_numeric(df[variable], errors='coerce')
    # Drop missing values before computing skewness
    data = df[variable].dropna()
    
    # Plot histogram
    sns.histplot(data, kde=True, ax=axes[i], label='Histogram')
    axes[i].set_title(f'Histogram of {variable}')
    
    # Check if there are still values after dropping missing values
    if len(data) > 0:
        # Calculate skewness, mean, and median
        skewness = data.skew()
        mean_value = data.mean()
        median_value = data.median()
        
        # Add information to legend
        legend_text = f'Skewness: {skewness:.2f}\nMean: {mean_value:.2f}\nMedian: {median_value:.2f}'
        axes[i].legend(title='Statistics', labels=[legend_text], loc='upper right', fontsize='small', handlelength=0)

plt.tight_layout()
plt.show()
```

| Column                        | Cleaning Operation                                            |
|-------------------------------|----------------------------------------------------------------|
| Age                           | NaN values replaced with median age, converted to integer      |
| Gender                        | NaN values replaced with 'Unknown', then 'Other' if still NaN  |
| Customer_ID                   | NaN values replaced with 'Unknown', appended with index if still NaN |
| Transaction_ID                | NaN values replaced with 'Unknown', appended with index if still NaN |
| Product_ID                    | NaN values replaced with 'Unknown', appended with index if still NaN |
| Income_Level                  | NaN values replaced with mode income level                    |
| Purchase_Date                 | NaN values replaced with mode purchase date                   |
| Product_Category              | NaN values replaced with mode product category                |
| Brand                         | NaN values replaced with mode brand                            |
| Purchase_Amount               | NaN values replaced with median purchase amount, converted to integer |
| Average_Spending_Per_Purchase | NaN values replaced with median average spending, converted to float |
| Purchase_Frequency_Per_Month  | NaN values replaced with median purchase frequency, converted to integer |
| Brand_Affinity_Score          | NaN values replaced with median brand affinity score, converted to float |
| Product_Category_Preferences   | NaN values replaced with mode product category preferences     |
| Month                         | NaN values replaced with mode month                            |
| Year                          | NaN values replaced with mode year                             |
| Season                        | NaN values replaced with mode season                           |
| Address                       | NaN values replaced with 'Address -> Unknown'                             |

```python
# replacing NaN values in Age Column with Median of Age
median_age = _raw_data['Age'].median()
_raw_data['Age'].fillna(median_age, inplace=True)
_raw_data['Age'] = _raw_data['Age'].astype(int)

# replacing NaN values in Gender Column with Unknow
_raw_data['Gender'].fillna('Unknown', inplace=True)
# _raw_data['Gender'].replace('Other','Unknown',inplace=True)

# replacing NaN values in Customer_ID Column with Unknow
_raw_data['Customer_ID'].fillna('Unknown', inplace=True)

# replacing NaN values in Transaction_ID Column with Unknow
_raw_data['Transaction_ID'].fillna('Unknown', inplace=True)

# replacing NaN values in Product_ID Column with Unknow
_raw_data['Product_ID'].fillna('Unknown', inplace=True)

# Replace NA values in Income_Level with the mode 
mode_income_level = _raw_data['Income_Level'].mode()[0]
_raw_data['Income_Level'].fillna(mode_income_level, inplace=True)

# Replace NA values in Purchase_Date with the mode 
mode_Purchase_Date = _raw_data['Purchase_Date'].mode()[0]
_raw_data['Purchase_Date'].fillna(mode_Purchase_Date, inplace=True)

# Replace NA values in Product_Catagory with the mode 
mode_Product_Catagory = _raw_data['Product_Category'].mode()[0]
_raw_data['Product_Category'].fillna(mode_Product_Catagory, inplace=True)

# Replace NA values in Brand with the mode 
mode_Brand = _raw_data['Brand'].mode()[0]
_raw_data['Brand'].fillna(mode_Brand, inplace=True)

# replacing NaN values in Purchase_Amount Column with Median of Purchase_Amount
median_Purchase_Amount = _raw_data['Purchase_Amount'].median()
_raw_data['Purchase_Amount'].fillna(median_Purchase_Amount, inplace=True)
_raw_data['Purchase_Amount'] = _raw_data['Purchase_Amount'].astype(int)

# replacing NaN values in Average_Spending_Per_Purchase Column with Median of Average_Spending_Per_Purchase
median_Average_Spending_Per_Purchase = _raw_data['Average_Spending_Per_Purchase'].median()
_raw_data['Average_Spending_Per_Purchase'].fillna(median_Average_Spending_Per_Purchase, inplace=True)
_raw_data['Average_Spending_Per_Purchase'] = _raw_data['Average_Spending_Per_Purchase'].astype(float)

# replacing NaN values in Purchase_Frequency_Per_Month Column with Median of Purchase_Frequency_Per_Month
median_Purchase_Frequency_Per_Month = _raw_data['Purchase_Frequency_Per_Month'].median()
_raw_data['Purchase_Frequency_Per_Month'].fillna(median_Purchase_Frequency_Per_Month, inplace=True)
_raw_data['Purchase_Frequency_Per_Month'] = _raw_data['Purchase_Frequency_Per_Month'].astype(int)

# replacing NaN values in Brand_Affinity_Score Column with Median of Brand_Affinity_Score
median_Brand_Affinity_Score = _raw_data['Brand_Affinity_Score'].median()
_raw_data['Brand_Affinity_Score'].fillna(median_Brand_Affinity_Score, inplace=True)
_raw_data['Brand_Affinity_Score'] = _raw_data['Brand_Affinity_Score'].astype(float)

# Replace NA values in Product_Category_Preferences with the mode 
mode_Product_Category_Preferences = _raw_data['Product_Category_Preferences'].mode()[0]
_raw_data['Product_Category_Preferences'].fillna(mode_Product_Category_Preferences, inplace=True)

# Replace NA values in Product_Category_Preferences with the mode 
mode_Month = _raw_data['Month'].mode()[0]
_raw_data['Month'].fillna(mode_Month, inplace=True)

# Replace NA values in Year with the mode 
mode_Year = _raw_data['Year'].mode()[0]
_raw_data['Year'].fillna(mode_Year, inplace=True)

# Replace NA values in Season with the mode 
mode_Season = _raw_data['Season'].mode()[0]
_raw_data['Season'].fillna(mode_Season, inplace=True)

#address by unknown
_raw_data['Address'].fillna('Address', inplace=True)

#customer ID
for i in range(len(_raw_data['Customer_ID'])):
    if _raw_data['Customer_ID'][i] == 'Unknown':
        _raw_data['Customer_ID'][i] = f'Unknown{i}'

#transaction_ID
for i in range(len(_raw_data['Transaction_ID'])):
    if _raw_data['Transaction_ID'][i] == 'Unknown':
        _raw_data['Transaction_ID'][i] = f'Unknown{i}'
    
#Product_ID
for i in range(len(_raw_data['Product_ID'])):
    if _raw_data['Product_ID'][i] == 'Unknown':
        _raw_data['Product_ID'][i] = f'Unknown{i}'
        
# Display information after cleaning
print("After Cleaning 🧹:")
print("\nDataFrame :")
print(_raw_data.info())
```

```python

# Create a dictionary to map income levels to randomly generated amounts
income_levels = {'High': np.random.randint(40000, 60000, size=1)[0],
                 'Medium': np.random.randint(20000, 39000, size=1)[0],
                 'Low': np.random.randint(10000, 19000, size=1)[0]}

# Function to apply randomly generated amounts based on income levels
def generate_random_amount(income_level):
    return np.random.randint(income_levels[income_level] - 5000, income_levels[income_level] + 5000, size=1)[0]

# Apply the function to create the 'Income_Level_Numeric' column
_raw_data['Income_Level_Numeric'] = _raw_data['Income_Level'].apply(generate_random_amount)

# Display the value counts of the 'Income_Level_Numeric' column
print(_raw_data['Income_Level_Numeric'].value_counts())
```

```python
# # Create new feature: Age Group
# bins = [0, 18, 30, 40, 50, 60, 100]
# labels = ['Under 18', '18-30', '31-40', '41-50', '51-60', '61+']
# _raw_data['Age_Group'] = pd.cut(_raw_data['Age'], bins=bins, labels=labels, right=False)

# _raw_data['Age_Group'].unique()


# function to create age groups
def createAgeGroup(age):
    if age <18:
        return 'Under 18'
    elif age <= 30:
        return '18-30'
    elif age <= 40:
        return '31-40'
    elif age <= 60:
        return '51-60'
    else:
        return 'Above 60'

# create a new age_group columns using exist ages
_raw_data['Age_Group'] = _raw_data['Age'].apply(createAgeGroup)
_raw_data['Age_Group'].unique()
```

```python


color_palette = sns.color_palette("colorblind")

# Plotting Age Distribution
plt.figure(figsize=(12, 6))
ax1 = sns.histplot(data=_raw_data, x='Age', bins=20, kde=True, color=color_palette[0])
plt.title('Customer Age Distribution')
plt.xlabel('Age')
plt.ylabel('Count')

ax1.bar_label(ax1.containers[0], fontsize=10, color='black', label_type='edge')
mean_age = _raw_data['Age'].mean()
skewness_age = _raw_data['Age'].skew()
variance_age = _raw_data['Age'].var()

ax1.text(0.95, 0.95, f'Mean: {mean_age:.2f}\nSkewness: {skewness_age:.2f}\nVariance: {variance_age:.2f}',
         transform=ax1.transAxes, ha='right', va='top', bbox=dict(boxstyle='round', facecolor='white', alpha=0.5))

plt.show()
```

## K_Means

### Select Age and Purchase_amount for clustering

- **Cluster Separation:** The clusters appear to be poorly separated, with significant overlap among data points. 🤔 This indicates that the clustering algorithm struggled to identify distinct groups of customers.

- **Relationship between Age and Purchase Amount:** The relationship between Age and purchase amount is not evident. 📉 There is no clear pattern suggesting that customers with higher Age tend to spend more. 

- **Centroid Information:** The centroids are not very informative.  They don't show significant variation, implying that the average data points for each cluster are similar. This further supports the notion that the clusters lack clear boundaries.

```python
X=_raw_data.iloc[:,[1,10]].values
_raw_data.info()
```

### Using the elbow method to find the optimal number of clusters

```python
from sklearn.cluster import KMeans
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters = i, init = 'random', random_state = 42)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)
plt.plot(range(1, 11), wcss,marker='o')
plt.title('The Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

### Training the K-Means model on the dataset

```python
kmeans = KMeans(n_clusters = 3, init = 'random', random_state = 42)
y_kmeans = kmeans.fit_predict(X)
```

### Visualising the clusters

```python
plt.scatter(X[y_kmeans == 0, 0], X[y_kmeans == 0, 1], s = 100, c = 'red', label = 'Cluster 1')
plt.scatter(X[y_kmeans == 1, 0], X[y_kmeans == 1, 1], s = 100, c = 'blue', label = 'Cluster 2')
plt.scatter(X[y_kmeans == 2, 0], X[y_kmeans == 2, 1], s = 100, c = 'green', label = 'Cluster 3')
# plt.scatter(X[y_kmeans == 3, 0], X[y_kmeans == 3, 1], s = 100, c = 'black', label = 'Cluster 4')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s = 300, c = 'yellow', label = 'Centroids')
plt.title('Clusters of customers')
plt.xlabel('Age')
plt.ylabel('Purchase_amount')
plt.legend()
plt.show()
```

### DB Scan
| **Aspect**                          | **1st DBSCAN**                                                                                                         | **2nd DBSCAN**                                                                                                       |
|-------------------------------------|---------------------------------------------------------------------------------------------------------------------------|--------------------------------------------------------------------------------------------------------------------------|
| **Data Preprocessing**              | No preprocessing applied to the data.                                                                                    | Features scaled using `StandardScaler` to ensure uniformity.                                                            |
| **Parameters**                      | `eps=0.2, min_samples=5`                                                                                                 | `eps=0.5, min_samples=5`                                                                                                 |
| **Visualization**                   | Basic scatter plot without edge color or scaling.                                                                        | Scatter plot with improved aesthetics, including edge color and standardized features.                                  |


DBSCAN (Density-Based Spatial Clustering of Applications with Noise) is a widely used clustering algorithm in machine learning for its ability to discover clusters of arbitrary shapes and handle noise within the data. Unlike K-Means, DBSCAN doesn't require specifying the number of clusters beforehand, making it particularly useful in scenarios where the data's underlying structure is unknown. It defines clusters based on the density of data points, effectively identifying areas of high density separated by regions of lower density.


```python
from sklearn.cluster import DBSCAN
# Apply DBSCAN
data=_raw_data.iloc[:,[1,10]].values
# Adjust the parameters (eps and min_samples) based on your dataset
dbscan = DBSCAN(eps=0.2, min_samples=5)
labels = dbscan.fit_predict(data)

# Plot the results
plt.scatter(data[:, 0], data[:, 1], c=labels, cmap='viridis')
plt.title('DBSCAN Clustering')
plt.xlabel('Age')
plt.ylabel('Purchase_Amount')
plt.legend()
```

```python
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
import pandas as pd

# Assuming '_raw_data' is your DataFrame
data = _raw_data.iloc[:, [1, 10]].values

# Standardize the data
scaler = StandardScaler()
data_scaled = scaler.fit_transform(data)

# Apply DBSCAN
dbscan = DBSCAN(eps=0.5, min_samples=5)
labels = dbscan.fit_predict(data_scaled)

# Plot the results
plt.scatter(data[:, 0], data[:, 1], c=labels, cmap='viridis', edgecolors='k', s=50)
plt.title('DBSCAN Clustering')
plt.xlabel('Age')
plt.ylabel('Purchase_Amount')
plt.show()
```
